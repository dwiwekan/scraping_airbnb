```python
import pandas as pd
import re
```

```python
# Read data
df_badung = pd.read_excel('Cleaning-Data/badung-airbnb.xlsx')
df_buleleng = pd.read_excel('Cleaning-Data/buleleng-airbnb.xlsx')
df_denpasar = pd.read_excel('Cleaning-Data/denpasar-airbnb.xlsx')
df_gianyar = pd.read_excel('Cleaning-Data/gianyar-airbnb.xlsx')
df_jembrana = pd.read_excel('Cleaning-Data/jembrana-airbnb.xlsx')
df_jembrana = pd.read_excel('Cleaning-Data/jembrana-airbnb.xlsx')
df_klungkung = pd.read_excel('Cleaning-Data/klungkung-airbnb.xlsx')
df_karangasem = pd.read_excel('Cleaning-Data/karangasem-airbnb.xlsx')
df_tabanan = pd.read_excel('Cleaning-Data/tabanan-airbnb.xlsx')
df_bangli = pd.read_excel('Cleaning-Data/bangli-airbnb.xlsx')
```

```python
df_gabung = pd.concat([df_badung, df_buleleng, df_denpasar, df_gianyar, df_jembrana, df_klungkung, df_karangasem, df_tabanan,df_bangli], ignore_index=True)
df_gabung
```

```python
df_gabung.drop_duplicates(subset ="link", keep = 'first', inplace = True)
df_gabung.dropna(subset=['judul', 'alamat', 'kategori'], inplace=True)
df_gabung.reset_index(drop=True, inplace=True)
df_gabung.info()
```

```python
# Menghilangkan kata "Bergabung sejak" dari kolom 'kalimat'
df_gabung['tahun'] = df_gabung['tahun'].str.replace('Bergabung sejak ', '')
df_gabung_indo = df_gabung[df_gabung['alamat'].str.contains('Indonesia')]
# df_gabung_indo.to_excel('Cleaning-Data/gabung-airbnb.xlsx', index=False)
```
