```python
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
import time
import re
import pandas as pd
import numpy as np
import datetime
```

```python
from selenium import webdriver
from parsel import Selector

chromedrive_path = './chromedriver' # use the path to the driver you downloaded from previous steps
driver = webdriver.Chrome(chromedrive_path)
url = 'https://www.balipost.com/pariwisata'
text = requests.get(url).text
driver.get(url)
# driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
try:
    input('Press any key to continue...')
    page_content = driver.page_source
    soup = BeautifulSoup(page_content, 'lxml')
    all_links = soup.find_all("a")
finally:
    driver.quit()
```

```python
# all_links
```

```python
link_list = list()
for idx,link in enumerate(all_links):
    try:
        judul = link.get('title')
        link = link.get('href')
        if "https://www.balipost.com/news/" in link:
            link_list.append((judul,link))
    except:
        continue

link_list = list(set(link_list))
```

```python
# link_list
```

```python
list_berita = list()
for idx,link in enumerate(link_list):
    date = re.search(r'\d{4}/\d{2}/\d{2}', link[1])
    date = date.group(0).replace("/", "-")
    date = datetime.datetime.strptime(date, '%Y-%m-%d').strftime('%d-%m-%Y')
    print(f'{idx} {date} {link[0]}')
    list_berita.append([date, link[0], link[1]])
```

```python
df = pd.DataFrame(list_berita, columns=['date', 'judul', 'link'])
df.to_excel('data_berita_bali.xlsx', index=False)
```
